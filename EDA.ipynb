```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clean_data = pd.read_csv('cleaned_gender_submission.csv')
```

```python
clean_data
```

```python
clean_data.describe()
```

```python
# Distribution plots for numeric variables
numeric_cols = clean_data.columns
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(clean_data[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

Relationships Between Variables

```python
# Correlation matrix
corr = clean_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for numeric relationships
sns.pairplot(clean_data[numeric_cols])
plt.show()
```

```python
categorical_cols = clean_data.select_dtypes(include='int64').columns
if len(categorical_cols) >= 2:
    ct = pd.crosstab(clean_data[categorical_cols[0]], clean_data[categorical_cols[1]])
    print("\nCross-tabulation:\n", ct)
    ct.plot(kind='bar', figsize=(10,5))
    plt.title(f"{categorical_cols[0]} vs {categorical_cols[1]}")
    plt.show()
```

 Patterns and Trends

```python
# Ensure correct column names
clean_data.columns = clean_data.columns.str.strip()

# Survival rate
survival_rate = clean_data['Survived'].value_counts(normalize=True)
print(survival_rate)

survival_rate.plot(kind='bar', figsize=(6,4))
plt.title("Survival Distribution")
plt.ylabel("Proportion")
plt.xlabel("Survived")
plt.show()
```

```python
clean_data.columns = clean_data.columns.str.strip()

# Survival rate
survival_rate = clean_data['PassengerId'].value_counts(normalize=True)
print(survival_rate)

survival_rate.plot(kind='bar', figsize=(6,4))
plt.title("PassengerId Distribution")
plt.ylabel("Proportion")
plt.xlabel("PassengerId")
plt.show()
```
